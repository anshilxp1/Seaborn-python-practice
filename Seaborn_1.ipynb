# Seaborn

```python
# we can visualize the multiple columns in the graph
# we can perform the aggregation function inside the graph
# we can perorm the statiscal analysis also
#
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
# Numerical plots
# - relplot
#     - line plot
#     - scatter plot
```

```python
sns.relplot(x='total_bill',y='tip',kind="line",data=df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',data=df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',data=df);
```

```python
df['size'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',style='size',data=df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',style='size',col='smoker',data=df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',style='size',row='time',col='smoker',data=df);
```

```python
sns.lineplot(x='total_bill',y='tip',data=df);
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df);
```

```python
# Categorical Plot

# Catplot
    # - swarmplot
    # - barplot
    # histplot
    # countplot
    # - boxplot
```

```python
sns.catplot(x='day',y='total_bill',hue='day',data=df);
```

```python
df['total_bill'].mean()
```

```python
sns.catplot(x='day',y='total_bill',hue='day',data=df); # strip plot
```

```python
sns.catplot(x='day',y='total_bill',hue='day',kind='swarm',data=df); # swarm plot
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df.groupby(['day'])['total_bill'].agg(['sum','min','max','mean','median'])
```

```python
sns.barplot(x='day',y='total_bill',estimator='max',ci=None ,data=df,palette='Set1');
```

```python
sns.catplot(x='day',y='total_bill',hue='day',kind='bar',ci=None,data=df); # swarm plot
```

```python
sns.histplot(x='size',data=df); # generally for numerical
```

```python
df['size'].value_counts()
```

```python
df['sex'].value_counts()
```

```python
sns.countplot(x='sex',hue='sex',data=df);
```

```python
sns.catplot(x='sex',kind='count',hue='sex',data=df); # count plot
```

```python
sns.barplot(x='day',y='total_bill',estimator='min',hue='sex',ci=None,data=df,palette='Set1');
```

```python
sns.boxplot(y='tip',hue='day',data=df);
```

```python
sns.boxplot(x='day',y='tip',hue='sex',data=df);
```
